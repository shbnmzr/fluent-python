# List Comprehensions and Generator Expressions

```python
symbols = '$!~£ù'
```

```python
codes = []

for symbol in symbols:
    codes.append(ord(symbol))

codes
```

```python
codes = [ord(symbol) for symbol in symbols]
codes
```

### Local Scope

```python
x = 'ABC'
codes = [ord(el) for el in x]
# el is not accessible outside the list comprehension
# Walrus operator defines the scope of the variable as the enclosing function
codes = [last := ord(el) for el in x]
last
```

### Listcomps vs. map and filter

```python
symbols = '$§&£~'

beyond_ascii = [ord(symbol) for symbol in symbols if ord(symbol) > 127]

beyond_ascii
```

```python
beyond_ascii = list(filter(lambda s: s > 127, map(ord, symbols)))
beyond_ascii
```

### Cartesian Product

```python
colors = ['black', 'white', 'blue']
sizes = ['S', 'M', 'L', 'XL']

tshirts = [(color, size) for color in colors
                        for size in sizes]

tshirts
```

```python
# Generator Expression
# A genexp (generator expression) saves memory because it yields items
# one by one using the iterator protocol instead of building a whole list just to feed
# another constructor.
symbols = '$§&£~'

tuple(ord(symbol) for symbol in symbols)
```

```python
import array

array.array('I', (ord(symbol) for symbol in symbols))
```

```python
colors = ['black', 'white', 'blue']
sizes = ['S', 'M', 'L', 'XL']

for tshirt in (f'{color} {size}' for color in colors for size in sizes):
    print(tshirt)
```

### Tuples

```python
a = (10, 'alpha', [1, 2])
b = (10, 'alpha', [1, 2])

print(a == b)
```

```python
b[-1].append(100)
print(b)

print(a == b)
```

```python
def fixed(obj):
    try:
        hash(obj)
    except TypeError:
        return False
    return True
```

```python
tf = (10, 'alpha', [1, 2])
tm = (10, 'alpha', (1, 2))

print(f'{tf} is hashable? {fixed(tf)}')
print(f'{tm} is hashable? {fixed(tm)}')
```
