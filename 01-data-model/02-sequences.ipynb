# List Comprehensions and Generator Expressions

```python
symbols = '$!~£ù'
```

```python
codes = []

for symbol in symbols:
    codes.append(ord(symbol))

codes
```

```python
codes = [ord(symbol) for symbol in symbols]
codes
```

### Local Scope

```python
x = 'ABC'
codes = [ord(el) for el in x]
# el is not accessible outside the list comprehension
# Walrus operator defines the scope of the variable as the enclosing function
codes = [last := ord(el) for el in x]
last
```

### Listcomps vs. map and filter

```python
symbols = '$§&£~'

beyond_ascii = [ord(symbol) for symbol in symbols if ord(symbol) > 127]

beyond_ascii
```

```python
beyond_ascii = list(filter(lambda s: s > 127, map(ord, symbols)))
beyond_ascii
```

### Cartesian Product

```python
colors = ['black', 'white', 'blue']
sizes = ['S', 'M', 'L', 'XL']

tshirts = [(color, size) for color in colors
                        for size in sizes]

tshirts
```

```python
# Generator Expression
# A genexp (generator expression) saves memory because it yields items
# one by one using the iterator protocol instead of building a whole list just to feed
# another constructor.
symbols = '$§&£~'

tuple(ord(symbol) for symbol in symbols)
```

```python
import array

array.array('I', (ord(symbol) for symbol in symbols))
```

```python
colors = ['black', 'white', 'blue']
sizes = ['S', 'M', 'L', 'XL']

for tshirt in (f'{color} {size}' for color in colors for size in sizes):
    print(tshirt)
```

### Tuples

```python
a = (10, 'alpha', [1, 2])
b = (10, 'alpha', [1, 2])

print(a == b)
```

```python
b[-1].append(100)
print(b)

print(a == b)
```

```python
def fixed(obj):
    try:
        hash(obj)
    except TypeError:
        return False
    return True
```

```python
tf = (10, 'alpha', [1, 2])
tm = (10, 'alpha', (1, 2))

print(f'{tf} is hashable? {fixed(tf)}')
print(f'{tm} is hashable? {fixed(tm)}')
```

### Unpacking

```python
t = (20, 8)

quotient, remainder = divmod(*t)
quotient, remainder
```

```python
a, b, *rest = range(10)
a, b, rest
```

```python
a, *rest, c, d = range(10)

a, rest, c, d
```

```python
def func(a, b, c, *rest):
    return a, b, c, rest

func(*[1, 2], 2, 3, 7, 8, *range(10))
```

```python
metro_areas = [
    ('Tokyo', 'JP', 36.933, (35.689722, 139.691667)),
    ('Delhi NCR', 'IN', 21.935, (28.613889, 77.208889)),
    ('Mexico City', 'MX', 20.142, (19.433333, -99.133333)),
    ('New York-Newark', 'US', 20.104, (40.808611, -74.020386)),
    ('São Paulo', 'BR', 19.649, (-23.547778, -46.635833)),
]
```

```python
print(f'{"":15} | {"latitude":>9} | {"longitude":>9}')
for name, _, _, (lat, lon) in metro_areas:
    if lon <= 0:
        print(f'{name:15} | {lat:4.9f} | {lon:4.9f}')
```

```python
print(f'{"":15} | {"latitude":>9} | {"longitude":>9}')
for record in metro_areas:
    match record:
        case [name, _, _, (lat, lon)] if lon <= 0:
            print(f'{name:15} | {lat:4.9f} | {lon:4.9f}')
```
